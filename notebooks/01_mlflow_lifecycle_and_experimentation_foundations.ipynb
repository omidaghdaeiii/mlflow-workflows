# 2. The Machine Learning Lifecycle

Machine learning follows the same principles as software engineering—but introduces additional complexity.


### Core Stages

1. **Model Engineering**
   - Feature design
   - Model selection
   - Training

2. **Model Evaluation**
   - Measuring performance
   - Comparing models

3. **Model Deployment**
   - Serving models in production


### Key Insight

Unlike traditional software:

> ML systems are **probabilistic**, not deterministic

This makes tracking and reproducibility significantly harder.

# 3. Real-World Challenges

### Challenges in the ML Lifecycle

In practice, teams face persistent issues:


##### Experiment Tracking

- Multiple models trained
- No structured comparison
- Results become fragmented


##### Reproducibility

- Code changes over time
- Environments differ
- Results cannot be replicated reliably


##### Deployment Complexity

- No standard packaging
- Difficult transitions from research → production


##### Core Problem

> ML development without structure leads to **unreliable systems**

# 4. Introducing MLflow

### What is MLflow?

MLflow is an **open-source platform** designed to manage the ML lifecycle.


##### What MLflow Provides

- Experiment tracking
- Reproducibility
- Standardized model packaging
- Centralized model management


##### Core Philosophy

> Make ML development **systematic, traceable, and reproducible**

# 5. MLflow Architecture

### MLflow Components

MLflow consists of four core components:


##### 1. Tracking

- Logs experiments
- Stores:
  - Metrics
  - Parameters
  - Artifacts


##### 2. Models

- Standard format for packaging models
- Enables consistent deployment


##### 3. Model Registry

- Central repository for models
- Supports:
  - Versioning
  - Staging (dev, staging, prod)


##### 4. Projects

- Standardizes ML code packaging
- Ensures reproducibility across environments


##### Mental Model

MLflow is not a tool—  
it is an **infrastructure layer for ML systems**

# Solution Explanation

### Solution — Designing a Proper MLflow Experiment

This solution demonstrates how to:

1. Create a well-structured experiment  
2. Apply meaningful, production-quality tags  
3. Set the experiment correctly  


### Design Principles Used

##### 1. Clear Naming Convention
Experiment names should:
- Be descriptive
- Reflect purpose
- Be consistent across projects


##### 2. Meaningful Tags
Tags should capture:

- Project context
- Ownership
- Stage of development
- Business relevance


##### 3. Reproducibility Mindset
Everything should be:
> Discoverable, searchable, and understandable later

# Solution Code

```python
import mlflow

# -----------------------------
# Step 1: Create Experiment
# -----------------------------

experiment_name = "CustomerChurn_Baseline_Model"

experiment_id = mlflow.create_experiment(experiment_name)

print("Experiment created with ID:", experiment_id)


# -----------------------------
# Step 2: Set Experiment
# -----------------------------

mlflow.set_experiment(experiment_name)


# -----------------------------
# Step 3: Add Tags
# -----------------------------

mlflow.set_experiment_tag("project", "Customer Churn Prediction")
mlflow.set_experiment_tag("owner", "Data Science Team")
mlflow.set_experiment_tag("stage", "baseline")
mlflow.set_experiment_tag("use_case", "classification")
mlflow.set_experiment_tag("priority", "high")

print("Experiment is set and tagged successfully.")
```
